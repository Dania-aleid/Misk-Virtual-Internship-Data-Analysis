```python
import pandas as pd

df = pd.read_excel(
    list(uploaded.keys())[0],
    sheet_name='Final_Dataset',
    engine='pyxlsb'
)
df.info()
```

```python
df['Column1'] = range(1, len(df) + 1)
```

```python
 df['program_desc'] = df.apply(
    lambda row: row['program_desc']
    if pd.notnull(row['program_desc'])
    else f"{row['program_genre']} {row['program_class']} - {row['original_name']}",
    axis=1
)
 df.info()
```

```python
df['program_desc'].isnull().sum()
```

```python
df_copy = df.copy()

summary = df_copy.groupby('program_name').agg(
    total_views = ('program_name', 'count'),
    total_duration = ('duration_seconds', 'sum'),
    unique_users = ('user_id_maped', 'nunique')
).reset_index()

summary_sorted = summary.sort_values(by='total_views', ascending=False)

top_10 = summary_sorted.head(10)

top_10
```

```python
top_10_with_class = top_10.merge(
    df_copy[['program_name', 'program_class']].drop_duplicates(),
    on='program_name',
    how='left'
)

top_10_with_class
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,6))

# Bar plot for total duration
plt.subplot(1, 2, 1)
sns.barplot(
    data=top_10_with_class,
    x='program_name',
    y='total_duration',
    hue='program_class',
    dodge=False
)
plt.xticks(rotation=45, ha='right')
plt.title('Total Watch Duration by Program')
plt.xlabel('Program Name')
plt.ylabel('Total Duration (seconds)')
plt.tight_layout()

# Bar plot for unique users
plt.subplot(1, 2, 2)
sns.barplot(
    data=top_10_with_class,
    x='program_name',
    y='unique_users',
    hue='program_class',
    dodge=False
)
plt.xticks(rotation=45, ha='right')
plt.title('Unique Users by Program')
plt.xlabel('Program Name')
plt.ylabel('Unique Users')
plt.tight_layout()

plt.show()
```

```python
hd_summary = df_copy.groupby('hd').agg(
    total_views = ('program_name', 'count'),
    total_duration = ('duration_seconds', 'sum'),
    unique_users = ('user_id_maped', 'nunique')
).reset_index()

hd_summary['quality'] = hd_summary['hd'].map({0: 'SD', 1: 'HD'})

hd_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
sns.barplot(data=hd_summary, x='quality', y='total_views')
plt.title('Total Views by Quality')

plt.subplot(1,3,2)
sns.barplot(data=hd_summary, x='quality', y='total_duration')
plt.title('Total Duration by Quality')

plt.subplot(1,3,3)
sns.barplot(data=hd_summary, x='quality', y='unique_users')
plt.title('Unique Users by Quality')

plt.tight_layout()
plt.show()
```
